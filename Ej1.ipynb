## Enunciado

Usamos la distribución conjunta dada para $X_1, X_2, X_3$ y calculamos: 
a) $P(X_1=1, X_2=0)$
b) $P(X_3=0)$
c) $P(X_2=1 | X_3=1)$
d) $P(X_1=0 | X_2=1, X_3=1)$
e) $P(X_1=0, X_2=1 | X_3=1)$

```python
# Distribución conjunta P(X1, X2, X3)
joint = {
    (0, 0, 0): 0.05,
    (1, 0, 0): 0.10,
    (0, 1, 0): 0.40,
    (1, 1, 0): 0.10,
    (0, 0, 1): 0.10,
    (1, 0, 1): 0.05,
    (0, 1, 1): 0.20,
    (1, 1, 1): 0.00,
}

p_a = sum(p for (x1, x2, x3), p in joint.items() if x1 == 1 and x2 == 0)
p_b = sum(p for (x1, x2, x3), p in joint.items() if x3 == 0)
p_c = (
    sum(p for (x1, x2, x3), p in joint.items() if x2 == 1 and x3 == 1)
    / sum(p for (x1, x2, x3), p in joint.items() if x3 == 1)
)
p_d = (
    sum(p for (x1, x2, x3), p in joint.items() if x1 == 0 and x2 == 1 and x3 == 1)
    / sum(p for (x1, x2, x3), p in joint.items() if x2 == 1 and x3 == 1)
)
p_e = (
    sum(p for (x1, x2, x3), p in joint.items() if x1 == 0 and x2 == 1 and x3 == 1)
    / sum(p for (x1, x2, x3), p in joint.items() if x3 == 1)
)

results_ej1 = {
    "a) P(X1=1, X2=0)": p_a,
    "b) P(X3=0)": p_b,
    "c) P(X2=1 | X3=1)": p_c,
    "d) P(X1=0 | X2=1, X3=1)": p_d,
    "e) P(X1=0, X2=1 | X3=1)": p_e,
}

for k, v in results_ej1.items():
    print(f"{k}: {v:.6f}")
```

## Resultados finales

- a) $P(X_1=1, X_2=0)=0.15$
- b) $P(X_3=0)=0.65$
- c) $P(X_2=1 | X_3=1)=0.571429$
- d) $P(X_1=0 | X_2=1, X_3=1)=1.0$
- e) $P(X_1=0, X_2=1 | X_3=1)=0.571429$
