## Enunciado

Red bayesiana en cadena: $R \to T \to L$.

Con las tablas: 
- $P(R)$
- $P(T|R)$
- $P(L|T)$

Calculamos: 
a) $P(T=1 | R=0)$
b) $P(T=1)$
c) $P(L=1 | R=0, T=1)$
d) $P(L=1 | T=1)$
e) $P(L=1 | R=0)$
f) $P(T=1 | L=0)$

```python
# R=1: llueve, R=0: no llueve
# T=1: hay tráfico, T=0: no hay tráfico
# L=1: llega tarde, L=0: no llega tarde

P_R = {1: 0.1, 0: 0.9}
P_T_given_R = {
    (1, 1): 0.8,
    (0, 1): 0.2,
    (1, 0): 0.1,
    (0, 0): 0.9,
}
P_L_given_T = {
    (1, 1): 0.3,
    (0, 1): 0.7,
    (1, 0): 0.1,
    (0, 0): 0.9,
}

p2_a = P_T_given_R[(1, 0)]
p2_b = sum(P_T_given_R[(1, r)] * P_R[r] for r in [0, 1])
p2_c = P_L_given_T[(1, 1)]
p2_d = P_L_given_T[(1, 1)]
p2_e = sum(P_L_given_T[(1, t)] * P_T_given_R[(t, 0)] for t in [0, 1])

P_L0 = sum(
    P_L_given_T[(0, t)] * sum(P_T_given_R[(t, r)] * P_R[r] for r in [0, 1])
    for t in [0, 1]
)
p2_f = P_L_given_T[(0, 1)] * p2_b / P_L0

results_ej2 = {
    "a) P(T=1 | R=0)": p2_a,
    "b) P(T=1)": p2_b,
    "c) P(L=1 | R=0, T=1)": p2_c,
    "d) P(L=1 | T=1)": p2_d,
    "e) P(L=1 | R=0)": p2_e,
    "f) P(T=1 | L=0)": p2_f,
}

for k, v in results_ej2.items():
    print(f"{k}: {v:.6f}")
```

## Resultados finales

- a) $P(T=1 | R=0)=0.10$
- b) $P(T=1)=0.17$
- c) $P(L=1 | R=0, T=1)=0.30$
- d) $P(L=1 | T=1)=0.30$
- e) $P(L=1 | R=0)=0.12$
- f) $P(T=1 | L=0)=0.137413$
